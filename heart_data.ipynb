```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="/content/heart_2020_cleaned.csv"
```

```python
df=pd.read_csv(path)
df
```

```python
df.info()
```

```python
#1
sns.displot(df["SleepTime"],kde=True,bins=5)
plt.title("Sleep Time Distribution")
```

Insights:-

Most people typically sleep between 6 to 10 hours per night






```python
#2
sns.countplot(x="Diabetic",data=df,hue="Sex",palette="flare")
plt.xticks(rotation=90)
plt.title("Diabetic Distribution of Males & Females")
```

Insights:-
As per the graph,whether male or female,most of them are non diabetic patients

```python
#3
sns.barplot(x="Stroke",y="BMI",data=df,color="red")
plt.title("Average BMI data of people With or Without Stroke")
```

Insights:-
As per the graph,the average BMI of those with stroke is high

```python
#4
plt.title("GenHealth vs PhysicalHealth")
sns.lineplot(x="GenHealth",y="PhysicalHealth",data=df,color="red",marker="*",markersize="10",markerfacecolor="y",estimator=np.mean)
```

Insights:-
Baised on this data it is seen that Whwn Genral Health is poor Average Physical Health is at its peak level

```python
#5
sns.violinplot(x="Sex",y="MentalHealth",data=df,hue="AlcoholDrinking",palette="rainbow")
```

Insights:-
As per the graph Among individuals with alcohol use disorder, mental health appears to be poorer in males compared to females.

```python
#6
plt.title("BMI Distribution")
sns.kdeplot(df["BMI"])
```

Insights:-
Most people have a BMI rate ranging between 15 and 45

```python
#7
sns.displot(df["SkinCancer"],color="red")
plt.title("SkinCancer Distribution")
```

Insights:-As per the data almost 3lack people doesnt have Skin Cancer

```python
#8
sns.barplot(x='Sex',y="BMI",data=df,hue="PhysicalActivity",palette="flare")
plt.legend(bbox_to_anchor=(1,1))
```

Insights:-
As per the graph people have highest BMI who have no physical activity,regardless of male or female

```python
#9
plt.style.use("dark_background")
heart_disease=df["HeartDisease"].value_counts()
plt.figure(figsize=(8,8))
color=["green","pink"]
plt.pie(heart_disease,labels=heart_disease.index,autopct="%1.1f%%",colors=color)
plt.title("HEART DISEASE PERCENTAGE BREAKDOWN")
plt.legend(bbox_to_anchor=(1,1))
```

Insights:-
The pie chart indicates that a significant majority,91% of the population does not have heart disease.

```python
#10
heart_disease1=df[df["HeartDisease"]=="Yes"]
gender=heart_disease1["Sex"].value_counts()
gender
```

```python
plt.title("Smoking Habits of Heart Patients by Gender")
sns.countplot(x="Sex",data=heart_disease1,hue="Smoking",palette="rainbow")
```

Insights:-
According to the Data,the number of Male Heart patients who smoke is above 10,000 compared to 6000 females.

```python
#11
plt.title("DISTRIBUTION OF HEARTDISEASE BY GENDER")
plt.pie(gender,labels=gender.index,autopct="%1.1f%%")
plt.legend(bbox_to_anchor=(1,1))
```

Insights:-
According to the Data,59% of people with heart disease are Male

```python
#12
#HRHD- HeartDisease Patients with high risk factors
HRHD=df[(df["Smoking"]=="Yes") & (df["HeartDisease"]=="Yes") & (df["AlcoholDrinking"]=="Yes") & (df["Diabetic"]=="Yes") & (df["PhysicalActivity"]=="No") & (df["BMI"]>30)]
high_risk=HRHD["Sex"].value_counts()
high_risk
```

```python
plt.title("ANALYSIS OF HIGH RISK HEART DISEASE")
color=["red","yellow"]
plt.pie(high_risk,labels=high_risk.index,colors=color,autopct="%1.1f%%")
plt.legend(bbox_to_anchor=(1,1))
```

Insights:-
The analysis reveals that 70% of heart patients with high-risk heart disease are male, compared to females

```python
#13
#HF-Healthy and fit
HF=df[(df["Smoking"]=="No") & (df["HeartDisease"]=="No") & (df["AlcoholDrinking"]=="No") & (df["SkinCancer"]=="No") & (df["Stroke"]=="No") & (df["Diabetic"]=="No") & (df["PhysicalActivity"]=="Yes") & (df["BMI"]<24) & (df["KidneyDisease"]=="No")]
HF["GenHealth"].value_counts()
```

```python
plt.title("Health of Risk-Free Individuals by Gender")
sns.countplot(x="Sex",data=HF,hue="GenHealth",palette="rainbow")
```

Insights:-
Among individuals without risk factors, over 2,000 females and more than 1,000 males are in excellent health

```python
#14
AgeCategory=df.groupby("AgeCategory")
Low_Age_group=AgeCategory.get_group("25-29")
LAHD=Low_Age_group["HeartDisease"].value_counts()
LAHD#low age group with heart disease
```

```python
plt.title("DISTRIBUTION OF HEART DISEASE IN AGE CATEGORY 25-29")
color=["cyan","yellow"]
plt.pie(LAHD,labels=LAHD.index,colors=color,autopct="%1.1f%%")
plt.legend(bbox_to_anchor=(1,1))
```

Insights:-
Among individuals in the age category of 25-29, 99.3% do not have cardiac conditions.



```python
#15
hd=df[df["HeartDisease"]=="Yes"]
hdc=hd['AgeCategory'].value_counts()
hdc
```

```python
plt.figure(figsize=(10,6))
plt.title("AGE DISTRIBUTION OF HEART PATIENTS")
plt.bar(hdc.index,hdc.values,color="brown")
plt.xlabel("AgeCategory")
plt.ylabel("No.of Heart patients")
plt.xticks(rotation=90)
plt.legend()
```

Insights:-
The majority of heart patients fall into the age group of 80 or older, with specifically over 1400 cases, as illustrated graphically.
